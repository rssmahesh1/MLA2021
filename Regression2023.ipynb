```python
import pandas as pd
data=pd.read_csv("Retail.csv")
data
```

```python
data['city'].unique()
```

```python
import statsmodels.api as sm
x = sm.add_constant( data['sales'] )
x.head(5)
```

```python
y=data['profit']
y
```

```python
model=sm.OLS(y, x).fit()
```

```python
model.summary2()
```

# Using categorical X variables

```python
x_features=['sales','discount','region','category']
```

```python
pd.get_dummies(data['region'])
```

```python
cat_features=['region','category']
```

```python
x_encoded=pd.get_dummies(data[x_features], columns=cat_features, drop_first=True)
```

```python
x_encoded.columns
```

```python
import statsmodels.api as sm
x = sm.add_constant(x_encoded)
x.head(5)
```

```python
model=sm.OLS(y, x).fit()
```

```python
model.summary2()
```
